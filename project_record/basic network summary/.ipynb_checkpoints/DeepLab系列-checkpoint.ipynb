## DeepLabv1和v2以及v3与v3+主要区别


[reference](https://towardsdatascience.com/review-deeplabv1-deeplabv2-atrous-convolution-semantic-segmentation-b51c5fbde92d )
    
1.	V1、V2均使用了Atrous Convolutions 和 Fullly connected Conditional Random Field(CRF); 
2.	相对V1，V2还提出了Atrous Spatial Pyramid Pooling (ASPP);
3.	V1使用VGGNet作为基础网络，V2则使用VGGNet和ResNet两者;


提出用CRF(conditional random field)细化分割结果；
提出用Atrous convolution来解决输出特征分辨率小的问题;
标准的卷积只能获取原图1/4的内容，而新的带孔卷积可以在全图上获取信息；
需要更小的upsample factor便可放大到原图像，使得分割结果更准确；

**Deeplabv1:**
DeepLabv1是在VGG16的基础上做了修改：
VGG16的全连接层转为卷积
最后的两个max池化层去掉了下采样
Pooling层后续卷积层的卷积核改为了空洞卷积
在ImageNet上预训练的VGG16权重上做finetune
引入CRF:
DCNN有分类定位步骤，CRF试图在像素之间的关系上建模；
临近的像素更可能具有相同的像素标签；
CRF考虑到一个像素分配给一个标签的概率；
通过迭代来优化像素的标记;

**DeeplabV2:**
用多尺度获得更好的分割效果(使用ASPP：Atrous Spatial Pyramid Pooling) 
基础层由VGG16转为ResNet
 
ASPP可以获得多尺度特征;
ASPP中在给定的Input Feature Map上以r=(6,12,18,24)的3×3空洞卷积并行采样；
ASPP将不同采样率得到的特征图最后融合在一起(相同通道做像素加操作)得到最后的预测结果；

•	模型优势：
速度上： 使用空洞卷积的Dense DCNN达到8fps，全连接的CRF需要0.5s
精准度：在几个先进的数据集上达到了先进的结果

**Deeplabv3:**

DeepLabv3是DeepLab rethinking后的结果，移除了post-processing step CRF;
DeepLabv3使用了 Multi-Grid操作；DeepLabv3在v2提出的ASPP基础上又添加了从Inceptionv2中借鉴的BN正则化项。
